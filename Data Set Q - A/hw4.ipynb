### Question 1

<p>For each programming skill level, compute:</p>
<ol>
<li><b>PropPythonOrJava</b>:the proportion (0 to 1, computed relatively to the group size) of people that know at least one between Python and Java. 
<li><b>StdevClass</b>:the standard deviation of the Classification knowledge (function <i>std</i>)
</ol>
<p>You should return one data frame with one row for each programming skill level and two columns named <b>PropPythonOrJava</b> and <b>StdevClass</b>. Don't worry about having a Hierarchical Index for the columns.</p> 

```python
df['PropPythonOrJava']=(df.Python == 1) | (df.Java== 1)
```

```python
df['PropPythonOrJava'].head(5)
```

```python
df.groupby('ProgSkills').agg({'PropPythonOrJava' : 'mean', 
                      'Classification' : 'std'}).rename(columns={'Classification' : 'StdevClass'})
```

### Question 3

For each level of Classification, return the rows of those people who, within their group, have the smallest value in the column <i>Job</i>.
